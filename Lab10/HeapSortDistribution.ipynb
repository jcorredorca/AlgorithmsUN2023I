# Experimental analysis of the Insertion Sort execution time

```
def insertion_sort(A):
    for j in range(1,len(A)):
        key = A[j]
        i = j - 1
        while (i >= 0) and (A[i] > key):
            A[i + 1] = A[i]
            i = i -1
        A[i + 1] =   key
```

HEAPSORT

```
def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[i] < arr[left]:
        largest = left

    if right < n and arr[largest] < arr[right]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)


def heapSort(arr):
    n = len(arr)

    # Construir un montículo máximo
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extraer elementos uno por uno del montículo
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr
```

# Insertion sort  time complexity histograms for permutations of $n$ different elements counting steps like CLRS I.2.1

$$t (n) =  2n - 1  + 3 \sum_{j=2,..,n} t_j $$

## Best case - increasingly sorted array
$$t_b(n) =  5n - 4 = \Theta(n)$$

## Worst case - de creasingly sorted array

$$t_w(n) = \frac{3}{2} n^2 + \frac{7}{2} n  - 4 = \Theta(n^2)$$

## Average case - expected number of comparisons + swaps on random unifrom permutation of $n$ different elements

$$t_a(n) = \frac{3}{4} n^2 + \frac{17}{4} n  - 4  = \Theta(n^2)$$



```
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1,21,1)

plt.plot(n, 5*n-4,  label = 'Insertion sort best time $5n-4$')
plt.plot(n, 3/4*n**2+17/4*n-4, label = 'Insertion average time $3/4 n*^2+17/4 n - 4$')
plt.plot(n, 3/2*n**2+7/2*n-4, label = 'Insertion average time $3/2 n*^2+7/2 n - 4$')
plt.xlabel('Array size')
plt.ylabel('Steps')
plt.legend(loc = 'upper left')
```

```
def insertion_sort_steps(A):
    steps = 0
    for j in range(1,len(A)):
        key = A[j]
        i = j - 1
        while (i >= 0) and (A[i] > key):
            A[i + 1] = A[i]
            i = i -1
            steps=steps+3
        steps=steps+1
        A[i + 1] = key
        steps=steps+4
    steps=steps+1
    return steps
```

```
def heapify(arr, n, i, steps):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[i] < arr[left]:
        largest = left

    if right < n and arr[largest] < arr[right]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        steps += 3
        steps = heapify(arr, n, largest, steps)

    return steps


def heap_sort_steps(arr):
    n = len(arr)
    steps = 0

    # Construir un montículo máximo
    for i in range(n // 2 - 1, -1, -1):
        steps = heapify(arr, n, i, steps)

    # Extraer elementos uno por uno del montículo
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        steps += 3
        steps = heapify(arr, i, 0, steps)

    steps += 1
    return steps
```

```
A = [1,2,3,4,5,6,7,8]
print(heap_sort_steps(A))
print(A)

A = [6,7,2,4,5,3,8,1]
print(heap_sort_steps(A))
print(A)

A = [8,7,6,5,4,3,2,1]
print(heap_sort_steps(A))
print(A)
```

Generate all permutations without repetitions of size *n* (for small *n*), plot the "exact" time  distribution (number of steps), and print the media and the standar deviation

```
from itertools import permutations
import matplotlib.pyplot as plt
import numpy as np
import math

n=3
X = range(n)
times = []

for P in permutations(X):
  t = heap_sort_steps(list(P))
  times.append(t)
b = np.array(range(min(times)-1, max(times)+1))+0.5
plt.hist(times, bins = b )
plt.title("Insetion Sort")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.show()

print("n",n)
print("number of permutations n! =",math.factorial(n))
print("exp min =",min(times))
print("min formula 5*n-4 =",5*n-4)
print("exp max =",max(times))
print("max formula 3/2*n**2+7/2*n-4 =",3/2*n**2+7/2*n-4)
print("exp mean =",np.mean(times))
print("mean formula 3/4*n**2+17/4*n-4 =",3/4*n**2+17/4*n-4)
print("stdev",np.std(times))
```

# Insertion sort  time complexity histograms for permutations of $n$ elements with repetitions counting steps like CLRS I.2.1

$$t (n) =  2n - 1  + 3 \sum_{j=2,..,n} t_j $$

## Best case - increasingly sorted array
$$t_b(n) =  5n - 4$$

## Worst case - de creasingly sorted array

$$t_w(n) = \frac{3}{2} n^2 + \frac{7}{2} n  - 4$$

## Average case - expected number of comparisons + swaps on random unifrom permutation of $n$ different elements

$$t_a(n) = \frac{3}{4}n^2 + \frac{7}{2}n -\frac{13}{4}= \Theta(n^2)$$

# Average case - expected number of comparisons + swaps on random unifrom permtation of $n$   elements with repetitions $t_a(n) = \frac{3}{4}n^2 + \frac{7}{2}n  - \frac{13}{4} = \Theta(n^2) $

```
# Average case on random uniform permtations with repetitions of  $n$   elements  $t_a(n) = \frac{3}{4}n^2 + \frac{7}{2}n -\frac{13}{4}= \Theta(n^2) $
import numpy as np
import pandas as pd
import scipy.linalg as la

n = np.array([3.0,4.0,5.0,6.0])    # Number n of sorted elments
t = np.array([14.0,22.75,33.0,44.75]) # Average comparisons + swaps for the corresponding n

data = pd.DataFrame({'n': n, 't': t})
data
```

```
X = np.column_stack([n**0,n**1,n**2,n**3])
a = la.solve(X,t)
a
```
